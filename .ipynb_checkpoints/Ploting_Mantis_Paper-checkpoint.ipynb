# KMeans Dataset Preprocessing

```python
from sklearn.cluster import KMeans

# Importing the dataset
# dataset = pd.read_csv('Mall_Customers.csv')
dataset = pd.read_csv('text.csv')
dataset.dtypes

# Build an array containing the two columns we interested in
X = dataset.iloc[:, [8,9]].values
#X1 = dataset.iloc[:,1].values

# Apply the Kmeans algorithm to the dataset
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(X)
```

```python
data8['Vehicle ID']
```

# Configure & Export Plot 2

```python
facet = sns.lmplot(data=data, x='Speed', y='SpeedDiff', hue='Vehicle ID', fit_reg=False, legend=False)
facet.ax.legend(loc=2)
plt.scatter(kmeans.cluster_centers_[0, 0], kmeans.cluster_centers_[0, 1], s=150, c='red', label='Low Traffic')
plt.scatter(kmeans.cluster_centers_[1, 0], kmeans.cluster_centers_[1, 1], s=150, c='blue', label='Congenstion')
plt.scatter(kmeans.cluster_centers_[2, 0], kmeans.cluster_centers_[2, 1], s=150, c='green', label='Normal Traffic')
plt.legend()
plt.savefig('plot2.png', dpi=300) 
```
